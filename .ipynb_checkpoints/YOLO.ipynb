<h2>YOLO

Retomaremos el proyecto por medio de otro método, el cual será con machine learning usando redes neuronales convolucionales; lo cual significa que el sistema “copia” por decirlo así, el mismo sistema que tienen las neuronas biológicas humanas y así, para lograr una visión por computadora seguir una serie de pasos: extracción de características, que se logra entrenando al sistema para identificar o detectar el objeto deseado; luego sigue la reducción por muestreo, donde el sistema logra identificar los objetos a los cuales fue entrenado y en este paso, no se requiere intervención humana; y por último, se realiza la clasificación final sobre las características extraídas.
Para esto, haremos uso del algoritmo YOLO (you only look once), el cual es un sistema de código abierto utilizado específicamente para detección de objetos, destacado por su gran velocidad de detección. YOLO sigue una serie de pasos para tener un resultado satisfactorio, los cuales son: 
1) Dividir la imagen n*n cuadrículas.
2) Calcula cada uno de los recuadros x veces para lograr así un bajo nivel de certidumbre a la hora de detectar el objeto.
3) Posteriormente, limita los recuadros y elimina a los que estén por debajo de este límite
Usaremos la implementación Darknet que es la oficial para el uso del algoritmo; además, usaremos la plataforma “Google Colaboratory”, que sirve para correr códigos de Python en cualquier navegador, maneja GPUs gratuitas, y por añadidura, Colab es un servicio de Jupyter notebook alojado, que usa el sistema operativo Linux.

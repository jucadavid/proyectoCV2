## Template matching

Leemos la imagen del template.

En primer lugar usamos un template obtenido de la manipulación de la misma imagen en la que posteriormente realizaríamos las pruebas

```python
template = cv2.cvtColor(cv2.imread("./templates/whiteKing.jpg"), cv2.COLOR_BGR2RGB)

plt.imshow(template)
```

Revisamos el tamaño del template asegurando que tuviera las dimensiones adecuadas para funcionar en el algoritmo de template matching

```python
print("Template Shape: ", template.shape)
```

Como se puede observar el template tiene un tamaño menor a la imagen original, por lo tanto puede ser ejecutado el algoritmo

```python
mapa = cv2.matchTemplate(img, template, cv2.TM_CCOEFF)

plt.imshow(mapa, cmap="gray")
```

Luego ejecutamos el algoritmo de template matching y obtuvimos un mapa de semejanza, donde tenemos valores más altos en los lugares en los que el template se parece más a la ventana de la imagen, y más bajos dónde éstos son más diferentes.

Como se puede observar se obtuvieron valores altos en muchos puntos del mapa (zonas claras). Por lo tanto no se puede identificar una zona específica donde el template se asemeje de manera suficientemente clara.

```python
h, w = template.shape[:-1]

maxValue = max(mapa.flatten())
loc = np.where( mapa >= maxValue-5)

for i in range(0, len(loc), 2):
    print("Point:", loc[i][0], loc[i+1][0])
    cv2.rectangle(img, (loc[i+1][0], loc[i][0]), (loc[i+1][0]+w, loc[i][0]+h), (0, 0, 255), 2)

fig, (p1, p2, p3) = plt.subplots(1, 3, figsize=(20, 5))
p1.imshow(img)
p2.imshow(template)
p3.imshow(mapa, cmap="gray")

plt.show()
```

Sin embargo, al utilizar un template obtenido de la misma imagen en la que ejecutamos el algoritmo. Pudimos identificar el punto en el que tiene una mayor similitud, hayando así el punto en el que se encuentra la pieza seleccionada.

Al realizar esta prueba pudimos identificar un patrón en el mapa de similitud, patrón que es muy similar a los cuadros del tablero. Por esta razón llegamos a la conclusión de que los cuadros que se pueden observar en el template estaban afectando bastante la identificación de la ventana adecuada.

```python
template = cv2.cvtColor(cv2.imread("./templates/Rey.jpg"), cv2.COLOR_BGR2RGB)

mapa = cv2.matchTemplate(img, template, cv2.TM_CCOEFF)

h, w = template.shape[:-1]

maxValue = max(mapa.flatten())
loc = np.where( mapa >= maxValue-5)

for i in range(0, len(loc), 2):
    print("Point:", loc[i][0], loc[i+1][0])
    cv2.rectangle(img, (loc[i+1][0], loc[i][0]), (loc[i+1][0]+w, loc[i][0]+h), (0, 0, 255), 2)

fig, (p1, p2, p3) = plt.subplots(1, 3, figsize=(25, 5))
p1.imshow(img)
p2.imshow(template)
p3.imshow(mapa, cmap="gray")

plt.show()
```

Por esta razón decidimos cambiar el fondo del template, poniendo un fondo rojo que no es semejante a ninguna ventana del tablero, solucionando de esta manera una parte del problema.

```python
img = cv2.imread("./test/685b860d412b91f5d4f7f9e643b84452_jpg.rf.2d78193e4021ae5ffb49ecd1060bebd7.jpg")

mapa = cv2.matchTemplate(img, template, cv2.TM_CCOEFF)

h, w = template.shape[:-1]

maxValue = max(mapa.flatten())
loc = np.where( mapa >= maxValue-5)

for i in range(0, len(loc), 2):
    print("Point:", loc[i][0], loc[i+1][0])
    cv2.rectangle(img, (loc[i+1][0], loc[i][0]), (loc[i+1][0]+w, loc[i][0]+h), (0, 0, 255), 2)

fig, (p1, p2, p3) = plt.subplots(1, 3, figsize=(25, 5))
p1.imshow(img)
p2.imshow(template)
p3.imshow(mapa, cmap="gray")

plt.show()
```

Al realizar esta corrección sobre el template nos dimos cuenta de que al usar una imagen en BGR, y teniendo el template en RGB, el template matching funciona de manera correcta para algunos casos. Sin embargo, cuando ambas imagenes comparten el mismo espacio de color el template matching no funciona.

Con este análisis concluimos que lo que puede estar pasando es que cuando el template y la imagen están en diferentes espacios de color (bgr y rgb), sólo se comparan de manera acertada los verdes en la imagen, mientras que los rojos y azules se truncan, dando como resultado la correcta identificación de la pieza, incluso en imágenes de las que no proviene el template.

```python
img = cv2.imread("./test/2f6fb003bb89cd401322a535acb42f65_jpg.rf.66c0a46773a9cd583fb96c3df41a9e0c.jpg")

mapa = cv2.matchTemplate(img, template, cv2.TM_CCOEFF)

h, w = template.shape[:-1]

maxValue = max(mapa.flatten())
loc = np.where( mapa >= maxValue-5)

for i in range(0, len(loc), 2):
    print("Point:", loc[i][0], loc[i+1][0])
    cv2.rectangle(img, (loc[i+1][0], loc[i][0]), (loc[i+1][0]+w, loc[i][0]+h), (0, 0, 255), 2)

img2 = cv2.imread("./test/d7887071e972604ddf5940d8eb2702e7_jpg.rf.5f20fe9a6c746d488d6d0478828478cb.jpg")

mapa2 = cv2.matchTemplate(img2, template, cv2.TM_CCOEFF)

h, w = template.shape[:-1]

maxValue = max(mapa2.flatten())
loc = np.where( mapa2 >= maxValue-5)

for i in range(0, len(loc), 2):
    print("Point:", loc[i][0], loc[i+1][0])
    cv2.rectangle(img2, (loc[i+1][0], loc[i][0]), (loc[i+1][0]+w, loc[i][0]+h), (0, 0, 255), 2)
    
fig, ((p1, p2), (p3, p4)) = plt.subplots(2, 2, figsize=(25, 5))
p1.imshow(img)
p2.imshow(mapa, cmap="gray")
p3.imshow(img2)
p4.imshow(mapa, cmap="gray")

plt.show()
```

Sin embargo, como podemos ver en esta prueba, éste método no funciona para todos los casos, especialmente cuando la pieza tiene una rotación diferente a la que se encuentra en el template, o cuándo tenemos varias piezas similares en un mismo tablero (como en este caso la reina).

```python
img = cv2.imread("./test/e0d38d159ad3a801d0304d7e275812cc_jpg.rf.16aa43d37a5a2d2c32402cfa877d0a0d.jpg")

mapa = cv2.matchTemplate(img, template, cv2.TM_CCOEFF)

h, w = template.shape[:-1]

maxValue = max(mapa.flatten())
loc = np.where( mapa >= maxValue-5)

for i in range(0, len(loc), 2):
    print("Point:", loc[i][0], loc[i+1][0])
    cv2.rectangle(img, (loc[i+1][0], loc[i][0]), (loc[i+1][0]+w, loc[i][0]+h), (0, 0, 255), 2)
    
fig, (p1, p2, p3) = plt.subplots(1, 3, figsize=(25, 5))
p1.imshow(img)
p2.imshow(template)
p3.imshow(mapa, cmap="gray")
```

Otro problema que nos encontramos al realizar la detección, fué en imágenes que no contenían la ficha que estabamos tratando de encontrar.
